```python
#import essential libraries
import keras_ocr as krs
import matplotlib.pyplot as plt
```

```python
# Pretrained weights for the detector and recognizer
pipeline=krs.pipeline.Pipeline()
```

```python
img_ocr=[
           krs.tools.read(url) for url in['https://image.shutterstock.com/image-photo/selective-focus-text-outside-wear-260nw-1931498282.jpg',
                                        'https://www.seekpng.com/png/detail/137-1376594_handwritten-2-free-rough-hand-letters-letter.png',
                                         
 
                                         ]
]
```

```python
len(img_ocr)
```

**Visualizing the Image**

```python
plt.figure(figsize=(10,10))
plt.imshow(img_ocr[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img_ocr[1])
```

**Prediction**

```python
predict = pipeline.recognize(img_ocr)
```

```python
fig, axs = plt.subplots(nrows=len(img_ocr), figsize=(15, 15))
for ax, image, predictions in zip(axs, img_ocr, predict):
    krs.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
